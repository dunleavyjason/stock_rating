```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_pickle("prices_fundamentals.pkl")
```

```python
df
```

```python
list(df.columns)
```

```python
df = df.sort_values(['ticker'], ascending=True) \
        .groupby(['ticker'], sort=False) \
        .apply(lambda x: x.sort_values(['date'], ascending=True)) \
        .reset_index(drop=True)
```

```python
df
```

```python
df["price_change"] = df.groupby("ticker")["close"].apply(pd.Series.pct_change)
```

```python
df
```

```python
df = df[df.price_change.notna()]
```

```python
daily_sector_price_change = pd.DataFrame(df.groupby(["date", "sector"])["price_change"].mean()).reset_index()
daily_sector_price_change
```

```python
df = pd.merge(df, daily_sector_price_change, how="left", on=["date", "sector"])
```

```python
df.rename(columns = {"price_change_x":"price_change", "price_change_y":"price_change_sector"}, inplace=True)
```

```python
df["price_change_norm_sector"] = df["price_change"] - df["price_change_sector"]
df
```

```python
daily_industry_price_change = pd.DataFrame(df.groupby(["date", "industry"])["price_change"].mean()).reset_index()
daily_industry_price_change
```

```python
df = pd.merge(df, daily_industry_price_change, how="left", on=["date", "industry"])
```

```python
df
```

```python
df.rename(columns = {"price_change_x":"price_change", "price_change_y":"price_change_industry"}, inplace=True)
```

```python
df["price_change_norm_industry"] = df["price_change"] - df["price_change_industry"]
df
```
