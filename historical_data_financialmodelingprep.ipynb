### Import DCFs

```python
def get_jsonparsed_data(url):
    """
    Receive the content of ``url``, parse it as JSON and return the object.

    Parameters
    ----------
    url : str

    Returns
    -------
    dict
    """
    response = urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)
```

```python
dcfs = []
```

```python
for ticker in core_tickers:
    url = ("https://financialmodelingprep.com/api/v3/historical-daily-discounted-cash-flow/"+ticker+"?limit=650&apikey=8c48192b57789b1b85a59db736780f87")
    dcf = get_jsonparsed_data(url)
    dcfs.append(dcf)
```

```python
from itertools import chain
dcf = pd.DataFrame(list(chain.from_iterable(dcfs)))
dcf
```

```python
dcf.to_pickle("dcf.pkl")
```
