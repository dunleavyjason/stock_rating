```python
def calendar_date(date):
    str_date = str(date)
    str_yr_mo_d = str_date.split(" ")[0].split("-")
    yr_mo_d = [int(i) for i in str_yr_mo_d]
    year = yr_mo_d[0]
    if yr_mo_d[1] <= 3:
        return str(year-1) + "-12-31"
    elif yr_mo_d[1] <= 6 and yr_mo_d[1] > 3:
        return str(year) + "-03-31"
    elif yr_mo_d[1] <= 9 and yr_mo_d[1] > 6:
        return str(year) + "-06-30"
    else:
        return str(year) + "-09-30"
```

```python
#enterprise value
prices_fundamentals["ev"] = prices_fundamentals["marketcap"] + prices_fundamentals["debt"] - prices_fundamentals["cashneq"]
```

```python
#enterprise value over EBIT
prices_fundamentals["evebit"] = prices_fundamentals["ev"] / prices_fundamentals["ebit"]
```

```python
#enterprise value over EBITDA
prices_fundamentals["evebit"] = prices_fundamentals["ev"] / prices_fundamentals["ebitda"]
```

```python
#price to book
prices_fundamentals["pb"] = prices_fundamentals["marketcap"] / prices_fundamentals["equity"]
```

```python
#price to earnings
prices_fundamentals["pe"] = prices_fundamentals["close"] / prices_fundamentals["eps"]

prices_fundamentals["pe"] = prices_fundamentals["pe"].replace([np.inf, -np.inf], 0)
```

```python
#peg - 1
prices_fundamentals["peg1"] = prices_fundamentals["pe"] / prices_fundamentals["eps_1y_pct"]

prices_fundamentals["peg1"] = prices_fundamentals["peg1"].replace([np.inf, -np.inf], 0)
```

```python
#peg - 2
prices_fundamentals["peg2"] = prices_fundamentals["pe"] / prices_fundamentals["eps_2y_pct"]

prices_fundamentals["peg2"] = prices_fundamentals["peg2"].replace([np.inf, -np.inf], 0)
```

```python
#peg - 3
prices_fundamentals["peg3"] = prices_fundamentals["pe"] / prices_fundamentals["eps_3y_pct"]

prices_fundamentals["peg3"] = prices_fundamentals["peg3"].replace([np.inf, -np.inf], 0)
```
