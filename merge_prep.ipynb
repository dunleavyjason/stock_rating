```python
import pandas as pd
import numpy as np
import pickle
```

### Import Quandl DataFrame

```python
quandl_df = pd.read_pickle("quandl_df.pkl")
quandl_df
```

### Import DCF Dataframe

```python
dcf = pd.read_pickle("dcf.pkl")
dcf.rename(columns={"symbol":"ticker"}, inplace=True)
dcf
```

```python
dcf["date"] = pd.to_datetime(dcf["date"])
```

```python
quandl_dcf = pd.merge(quandl_df, dcf, on=["ticker", "date"], how='inner')
quandl_dcf
```

### Import Earnings Surprises DataFrame

```python
surprise = pd.read_pickle("surprise.pkl")
```

```python
surprise["date"] = pd.to_datetime(surprise["date"])
```

```python
surprise["earnings_surprise"] = surprise["actualEarningResult"] - surprise["estimatedEarning"]
```

```python
surprise.rename(columns={"symbol":"ticker"}, inplace=True)
```

```python
surprise
```

```python
quandl_dcf_surprise = pd.merge(quandl_dcf, surprise, on=["date", "ticker"], how="left")
quandl_dcf_surprise
```

```python
quandl_dcf_surprise[['actualEarningResult', 'estimatedEarning', 'earnings_surprise']] = quandl_dcf_surprise.groupby("ticker")\
    ['actualEarningResult', 'estimatedEarning', 'earnings_surprise']\
    .transform(lambda x: x.fillna(method='ffill'))
```

```python
quandl_dcf_surprise = quandl_dcf_surprise[quandl_dcf_surprise["earnings_surprise"].notna()]
```

```python
quandl_dcf_surprise
```

```python
quandl_dcf_surprise.reset_index(inplace=True, drop=True)
quandl_dcf_surprise
```

### Investigate and Export Final Dataframe

```python
final_df = quandl_dcf_surprise
```

```python
final_df
```

```python
final_df.info(verbose=True, null_counts=True)
```

```python
final_df.to_pickle("final_df.pkl")
```

```python
final_df.to_csv("final_df.csv")
```

### Export List of Final Tickers for Streamlit App

```python
ticker_name = final_df[["ticker", "name"]]
ticker_name["tick_name"] = ticker_name.ticker + " - " + ticker_name.name
```

```python
ticker_name
```

```python
final_tickers = list(ticker_name.tick_name.unique())
```

```python
with open('final_tickers.pkl', 'wb') as handle:
    pickle.dump(final_tickers, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
